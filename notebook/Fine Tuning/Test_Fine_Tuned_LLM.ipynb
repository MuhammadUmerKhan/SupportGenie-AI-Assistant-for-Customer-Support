## **🔐 2. Import Libraries and Login to Hugging Face**

```python
import os
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, BitsAndBytesConfig
from huggingface_hub import login

login()  # Paste your Hugging Face token when prompted
```

## **⚙️ 3. Configure 4-bit Quantization**

```python
# Use BitsAndBytes for quantized loading to save memory
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=False,
)
```

## **🔤 4. Load Tokenizer**

```python
tokenizer = AutoTokenizer.from_pretrained("Muhammad-Umer-Khan/Mistral-7b-v03-FAQs-Finetuned")
```

## **🧠 5. Load Fine-Tuned Mistral-7B Model (4-bit)**

```python
model = AutoModelForCausalLM.from_pretrained(
    "Muhammad-Umer-Khan/Mistral-7b-v03-FAQs-Finetuned",
    quantization_config=bnb_config,
    device_map="auto",
    trust_remote_code=True
)
```

## **🧪 6. Create Inference Pipeline**

```python
# Create inference pipeline
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=200,
    device_map="auto"
)
```

## **💬 7. Define Chatbot Response Function**

```python
def get_chatbot_response(user_input):
    try:
        prompt = f"<s>[INST] {user_input} [/INST]"
        result = pipe(prompt)[0]['generated_text']
        return result.replace(prompt, '').strip()
    except Exception as e:
        return "Sorry, I encountered an issue while generating the response."
```

## **📥 8. Test the Chatbot**

```python
get_chatbot_response("How can I change the password?")
```
